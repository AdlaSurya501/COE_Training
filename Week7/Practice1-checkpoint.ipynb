```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]


numeric_columns = df.select_dtypes(include=['number']).columns
for col in numeric_columns:
    df = remove_outliers(df, col)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder

df = df.copy()

df = df.drop(columns=["Last_Updated"], errors='ignore')

target_column = "Daily_Consumption"  # Replace with the actual target column name


x = df.drop(columns=[target_column])  # Features
y = df[target_column]  # Target variable

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=40)

clf = RandomForestClassifier(n_estimators=100, random_state=45)
clf.fit(X_train, Y_train)

Y_pred = clf.predict(X_test)

accuracy = accuracy_score(Y_test, Y_pred)
print(f"Accuracy: {accuracy:.4f}")

print("Classification Report:")
print(classification_report(Y_test, Y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Define probability of success
p = 0.6

# Generate Bernoulli samples
data = bernoulli.rvs(p, size=1000)

# Plot the distribution
plt.bar([0, 1], [sum(data == 0), sum(data == 1)], color=['pink', 'black'], tick_label=["Failure (0)","Success (1)"])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.xticks([0, 1], labels=["Failure (0)", "Success (1)"])
plt.title("Bernoulli Distribution (p=0.6)")
plt.show()
```

```python
from scipy.stats import binom
n = 10  
p = 0.3 

x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)

plt.bar(x, pmf, color='black', alpha=0.7)
plt.xlabel("Number of Successes (Clicks)")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lambda_val = 5

x = np.arange(0, 10)  
pmf = poisson.pmf(x, lambda_val)

plt.bar(x, pmf, color='grey', alpha=0.7)
plt.xlabel("Number of Calls per Minute")
plt.ylabel("Probability")
plt.title(f"Poisson Distribution (Î»={lambda_val})")
plt.show()
```

```python
from scipy.stats import norm

mu = 70
sigma = 10

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf, color='red', label=f"N({mu}, {sigma}^2)")
plt.xlabel("Range Of scores")
plt.ylabel("Probability Density")
plt.title(f"Normal Distribution (Î¼={mu}, Ïƒ={sigma})")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

z = np.linspace(-4, 4, 1000)

pdf = norm.pdf(z, 0, 1)

plt.plot(z, pdf, color='blue', label="Standard Normal Distribution (Î¼=0, Ïƒ=1)")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.title("Standard Normal Distribution (Z-Distribution)")
plt.axvline(x=0, color='red', linestyle='--', label="Mean (Z=0)")
plt.legend()
plt.grid()
plt.show()
```
